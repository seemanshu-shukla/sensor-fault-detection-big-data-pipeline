```python
from typing import List
import pandas as pd


class Sensor:


    def __init__(self, record: dict):
        for k, v in record.items():
            setattr(self, k, v)

    @staticmethod
    def dict_to_object(data: dict, ctx):
        print(data, ctx)
        return Sensor(record=data)

    def to_dict(self):
        return self.__dict__
    @classmethod
    def get_sensor_record(cls, file_path):
        chunk_df = pd.read_csv(file_path,chunksize=10)
        n_row = 0
        for df in chunk_df:
            for data in df.values:
                sensor = Sensor(dict(zip(df.columns, data)))
                # cars.append(car)
                #print(n_row)
                n_row += 1
                yield sensor

    @classmethod
    def get_schema(cls,file_path):
        columns = next(pd.read_csv(file_path,chunksize=10)).columns
    
        schema = dict()
        schema.update({
            "$id": "http://example.com/myURI.schema.json",
            "$schema": "http://json-schema.org/draft-07/schema#",
            "additionalProperties": "false",
            "description": "Sample schema to help you get started.",
            'properties': dict(),
            "title": "SampleRecord",
            "type": "object"})
        for column in columns:
            schema['properties'].update(
                {
                    column: {
                        "description": f"sensor {column} ",
                        "type": "string"
                    }
                }
            )
        return schema

    def __str__(self):
        return f"{self.record}"

```

# How to set the class attributes defined within constructor dynamically.

When we write self.sensor1 we set the attribures statically. lets see how can we do the same dynamically

```python
class Emp:
    def __init__(self,record:dict):
        for att_name,values in record.items():
            setattr(self,att_name,values) ##Here is how we are setting attributes dynamically
    
    @classmethod ##for class method refer to this link: https://www.geeksforgeeks.org/classmethod-in-python/
    def get_object(cls,record):
        return Emp(record)
```

```python
import pandas as pd
emp_info = {"name":["Seemanshu","Shukla"], "designation":["Data","Scientist"]}
df=pd.DataFrame(emp_info)
df
```

##### Converting the dataframe into dict()

```python
# For converting dataframe into dict() we will be using zip() method
columns = df.columns
print(columns)
print(df.values)  #Please note that here we have called .values over dataframe and not dict()
```

```python
for data in df.values:  #This gives each record/row in a separate list
    print(dict(zip(columns,data)))

## Observe the output we have finally converted the dataframe into dict() using zip method.
```

##### Converting the dict() into objects

```python
## For converting dict() into objects we will be using class since object is an instance of the class so we will
## definetly be requiring a class to which our dict() records must be referring

class Emp:
    def __init__(self,record:dict):
        for att_name,values in record.items():
            setattr(self,att_name,values) ##Here is how we are setting attributes dynamically
    
    @classmethod ##for class method refer to this link: https://www.geeksforgeeks.org/classmethod-in-python/
    def get_object(cls,record):
        return Emp(record)

for data in df.values:  #This gives each record/row in a separate list
    record = dict(zip(columns,data)) #dict() record for each row
    print(Emp.get_object(record))  #Observe the outputs we have converted the dict() into respective objects
```

## Demo use of yield and return statement inside function's body

Follow this link to understand the difference: https://www.geeksforgeeks.org/difference-between-yield-and-return-in-python/

##### In short when we want to send the result imediately to the caller without waiting for results of other iterations in queue(Property of a generator function) then we will be using yeild. Whereas, cases where we want to send the consolidated result of all the iteration at once(meaning waiting until all the iterations in queue have executed successsfully) to the caller function then in such cases we would be using return

```python
HYPEN_E_DOT = "-e ."

from typing import List
def get_requirements(file_path:str)->List[str]:
    """
    This function will return list of requirements
    """
    requirement_list:List[str] = []

    """
    Write a code to read requirements.txt file and append each requirements in requirement_list variable.
    """
    with open(file_path) as file_obj:
        requirement_list=file_obj.readlines()
        requirement_list=[req.replace("\n","") for req in requirement_list]

        if HYPEN_E_DOT in requirement_list:
            requirement_list.remove(HYPEN_E_DOT)

    return requirement_list

x = get_requirements("requirements.txt")
type(x)
```
